```python
import os
import zipfile
import cv2
import numpy as np
import pandas as pd
import string
from sklearn.model_selection import train_test_split
from tqdm import tqdm

# 1. Dezarhivează arhiva asl_alphabet_train.zip în folderul curent
zip_path = "asl_alphabet_train.zip"
extract_path = "asl_alphabet_train"
with zipfile.ZipFile(zip_path, "r") as zip_ref:
    zip_ref.extractall(extract_path)

# 2. Definim etichetele valide (litere A–Z)
VALID_LABELS = set(string.ascii_uppercase)

# 3. Funcție care scanează un director și construiește DataFrame cu (filepath, label)
def build_df_from_dir(root_dir):
    records = []
    for label in sorted(os.listdir(root_dir)):
        if label not in VALID_LABELS:
            continue
        folder = os.path.join(root_dir, label)
        if not os.path.isdir(folder):
            continue
        for fname in os.listdir(folder):
            if fname.lower().endswith(('.jpg', '.jpeg', '.png')):
                fullpath = os.path.join(folder, fname)
                records.append({'filepath': fullpath, 'label': label})
    return pd.DataFrame(records)

# 4. Construim DataFrame-ul din folderul dezarhivat
base_dir = os.path.join(extract_path, "asl_alphabet_train")  # probabil structura e asl_alphabet_train/asl_alphabet_train/A/...
if not os.path.isdir(base_dir):  # fallback dacă nu există subfolder
    base_dir = extract_path

df_all = build_df_from_dir(base_dir)
df_all = df_all.sample(frac=1, random_state=42).reset_index(drop=True)

print("Coloane df_all:", df_all.columns.tolist())
print("Primele 5 intrări:\n", df_all.head())
print("\nDistribuție per literă (A–Z):\n", df_all['label'].value_counts().head())
print("\nNumăr total imagini (A–Z):", len(df_all))

# 5. Împărțire stratificată: 70% train, 15% val, 15% test
df_train, df_temp = train_test_split(
    df_all,
    test_size=0.30,
    stratify=df_all['label'],
    random_state=42
)
df_val, df_test = train_test_split(
    df_temp,
    test_size=0.50,
    stratify=df_temp['label'],
    random_state=42
)

print(f"\nDimensiuni split-uri → Train: {len(df_train)}, Val: {len(df_val)}, Test: {len(df_test)}")

# 6. Salvăm CSV-urile cu split-urile
df_train.to_csv('split_train.csv', index=False)
df_val.to_csv('split_val.csv', index=False)
df_test.to_csv('split_test.csv', index=False)

# 7. Funcție de segmentare bazată pe skin‐color thresholding în spațiul YCrCb
def segment_hand_skin(input_path, crop_size=224):
    img = cv2.imread(input_path)
    if img is None:
        return None
    img_ycrcb = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)
    mask = cv2.inRange(img_ycrcb, np.array([0, 133, 77]), np.array([255, 173, 127]))
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
    mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel, iterations=2)
    mask = cv2.morphologyEx(mask, cv2.MORPH_DILATE, kernel, iterations=1)
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    if not contours:
        return None
    c = max(contours, key=cv2.contourArea)
    (x, y, w, h) = cv2.boundingRect(c)
    margin = 20
    x1 = max(x - margin, 0)
    y1 = max(y - margin, 0)
    x2 = min(x + w + margin, img.shape[1])
    y2 = min(y + h + margin, img.shape[0])
    crop = img[y1:y2, x1:x2]
    if crop.size == 0:
        return None
    crop_resized = cv2.resize(crop, (crop_size, crop_size))
    return crop_resized

# 8. Funcție care salvează crop-urile segmentate
def process_and_save_segmented(df, subset_name, output_base='segmented', crop_size=224):
    for _, row in tqdm(df.iterrows(), total=len(df), desc=f"Segmentare {subset_name}"):
        src_path = row['filepath']
        label = row['label']
        filename = os.path.basename(src_path)
        cropped = segment_hand_skin(src_path, crop_size=crop_size)
        if cropped is None:
            continue
        dest_dir = os.path.join(output_base, subset_name, label)
        os.makedirs(dest_dir, exist_ok=True)
        dest_path = os.path.join(dest_dir, filename)
        cv2.imwrite(dest_path, cropped)

# 9. Rulăm procesarea pentru fiecare subset
process_and_save_segmented(df_train, 'train', output_base='segmented', crop_size=224)
process_and_save_segmented(df_val,   'val',   output_base='segmented', crop_size=224)
process_and_save_segmented(df_test,  'test',  output_base='segmented', crop_size=224)

# 10. Verificăm câte imagini au fost salvate, pe subset și pe literă
for subset in ['train', 'val', 'test']:
    total_count = 0
    print(f"\n=== {subset.upper()} ===")
    subset_dir = os.path.join('segmented', subset)
    if not os.path.isdir(subset_dir):
        print(f"Nu există {subset_dir}")
        continue
    for letter in sorted(os.listdir(subset_dir)):
        letter_dir = os.path.join(subset_dir, letter)
        if not os.path.isdir(letter_dir):
            continue
        count = len([f for f in os.listdir(letter_dir) if f.lower().endswith(('.jpg', '.jpeg', '.png'))])
        print(f"{letter}: {count} imagini crop-uite")
        total_count += count
    print(f"Total {subset}: {total_count} imagini\n")
```

```python
import mediapipe as mp
import cv2
import pandas as pd
import numpy as np
import os
from tqdm import tqdm

mp_hands = mp.solutions.hands

def extract_hand_keypoints(image_path):
    img = cv2.imread(image_path)
    if img is None:
        return None
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    with mp_hands.Hands(static_image_mode=True, max_num_hands=1) as hands:
        results = hands.process(img_rgb)
        if not results.multi_hand_landmarks:
            return None
        # 21 keypoints * 3 coordonate (x,y,z)
        kp = []
        for lm in results.multi_hand_landmarks[0].landmark:
            kp.extend([lm.x, lm.y, lm.z])
        return kp

def process_segmented_to_keypoints(seg_base='segmented'):
    all_records = []
    for subset in ['train', 'val', 'test']:
        subset_dir = os.path.join(seg_base, subset)
        if not os.path.isdir(subset_dir):
            continue
        for label in sorted(os.listdir(subset_dir)):
            label_dir = os.path.join(subset_dir, label)
            if not os.path.isdir(label_dir):
                continue
            for fname in tqdm(os.listdir(label_dir), desc=f"{subset}/{label}"):
                if not fname.lower().endswith(('.jpg', '.jpeg', '.png')):
                    continue
                fpath = os.path.join(label_dir, fname)
                kp = extract_hand_keypoints(fpath)
                if kp is not None:
                    all_records.append({
                        "subset": subset,
                        "label": label,
                        "filename": fname,
                        **{f"kpt_{i}": v for i, v in enumerate(kp)}
                    })
    return pd.DataFrame(all_records)

df_kp = process_segmented_to_keypoints()
df_kp.to_csv("mediapipe_hand_keypoints.csv", index=False)
print("Shape:", df_kp.shape)
print("Columns:", df_kp.columns)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import joblib

# Pregătim datele
df_train = df_kp[df_kp["subset"] == "train"]
df_val   = df_kp[df_kp["subset"] == "val"]
df_test  = df_kp[df_kp["subset"] == "test"]

X_train = df_train[[c for c in df_train.columns if c.startswith("kpt_")]].values
y_train = df_train["label"].values
X_val   = df_val[[c for c in df_val.columns if c.startswith("kpt_")]].values
y_val   = df_val["label"].values
X_test  = df_test[[c for c in df_test.columns if c.startswith("kpt_")]].values
y_test  = df_test["label"].values

# Antrenare model
clf = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
clf.fit(X_train, y_train)

# Metrici pe test
y_pred = clf.predict(X_test)
print("Acuratețe test:", accuracy_score(y_test, y_pred))
print("\nClassification report:\n", classification_report(y_test, y_pred))
print("\nConfusion matrix:\n", confusion_matrix(y_test, y_pred))

# Salvare model și descărcare
joblib.dump(clf, "rf_mediapipe_hands.pkl")

from google.colab import files
files.download("rf_mediapipe_hands.pkl")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay

# 1. Matrice de confuzie colorată
plt.figure(figsize=(14, 12))
cm = confusion_matrix(y_test, y_pred, labels=clf.classes_)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=clf.classes_, yticklabels=clf.classes_)
plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.title('Confusion Matrix')
plt.tight_layout()
plt.savefig("confusion_matrix.png")
plt.show()

# 2. Classification report ca tabel imagine
report_dict = classification_report(y_test, y_pred, labels=clf.classes_, output_dict=True)
report_df = pd.DataFrame(report_dict).transpose()
plt.figure(figsize=(10, 8))
sns.heatmap(report_df.iloc[:-3, :-1], annot=True, cmap='YlGnBu', fmt=".2f")
plt.title("Classification Report (precision, recall, f1-score, support)")
plt.xlabel("Metrics")
plt.ylabel("Class")
plt.tight_layout()
plt.savefig("classification_report.png")
plt.show()

# 3. Distribuția predicțiilor pe clase (barplot)
plt.figure(figsize=(12, 4))
pred_counts = pd.Series(y_pred).value_counts().sort_index()
true_counts = pd.Series(y_test).value_counts().sort_index()
width = 0.35
ind = np.arange(len(clf.classes_))
plt.bar(ind - width/2, true_counts[clf.classes_], width, label='True')
plt.bar(ind + width/2, pred_counts[clf.classes_], width, label='Predicted')
plt.xticks(ind, clf.classes_)
plt.xlabel("Class")
plt.ylabel("Count")
plt.title("Distribuție: True vs Predicted Classes")
plt.legend()
plt.tight_layout()
plt.savefig("class_distribution.png")
plt.show()

# 4. Descărcare imagini rezultate (Colab)
from google.colab import files
files.download("confusion_matrix.png")
files.download("classification_report.png")
files.download("class_distribution.png")
```
